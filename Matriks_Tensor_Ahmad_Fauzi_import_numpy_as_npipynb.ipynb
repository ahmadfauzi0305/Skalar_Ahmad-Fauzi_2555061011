```python
import numpy as np
```

```python
# buat matrix X berukuran 2x3 menggunakan numpy
x = np.array([[30, 2, 5], [4, 7, 11],])
x
```

```python
# tampilkan ukuran dari matrix x
x.shape
```

```python
# tampilkan seluruh elemen dari kolom pertama dari matrix X
x[0:,]
```

```python
# tampilkan seluruh elemen dari baris kedua dari matrix X
x[:,1]
```

```python
import torch
```

```python
# buat matrix X berukuran 2x3 menggunakan pytorch
x_pt = torch.tensor([[30, 12, 15], [6, 10, 11],])
x_pt
```

```python
# tampilkan ukuran dari matrix x
x_pt.shape
```

```python
# tampilkan seluruh elemen dari kolom pertama dari matrix X
x_pt[0,:]
```

```python
# tampilkan seluruh elemen dari baris kedua dari matrix X

x_pt[:,1]
```

```python
import tensorflow as tf
```

```python
# buat matrix X berukuran 2x3 menggunakan tensorflow
x_tf = tf.Variable([[30, 12, 15], [6, 10, 11],])
x_tf
```

```python
# tampilkan ukuran dari matrix x
tf.shape(x_tf)
```

```python
# tampilkan seluruh elemen dari kolom pertama dari matrix X
x_tf[0,:]
```

```python
# tampilkan seluruh elemen dari baris kedua dari matrix X
x_tf[:,1]
```
